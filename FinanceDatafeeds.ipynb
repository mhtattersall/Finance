```python
stocks = ["AMZN","MSFT","INTC","GOOG","INFY.NS","3988.HK"] #.hs indian, .hk is hong kong
start = dt.datetime.today()-dt.timedelta(30) #using datetime, going back 30 days
end = dt.datetime.today()
cl_price = pd.DataFrame() # empty dataframe which will be filled with closing prices of each stock
ohlcv_data = {} # empty dictionary which will be filled with ohlcv dataframe for each ticker
```

```python
# looping over tickers and creating a dataframe with close prices
for ticker in stocks:
    cl_price[ticker] = yf.download(ticker,start,end)["Adj Close"] #we just want the adjusted close price
print(cl_price.head(10))
```

```python
# looping over tickers and storing OHLCV dataframe in dictionary
for ticker in stocks:
    ohlcv_data[ticker] = yf.download(ticker,start,end) #dictionary used to store the date as key and df with all ohlcv with as the value
print(len(ohlcv_data)) #returns the number of keys i.e. 1 for each stock
```

```python
#to access the prices in the dictionary:
ohlcv_data["MSFT"]
ohlcv_data["MSFT"]["Open"]
```

```python
#alternative to yfinance feed.  Uses webscraping, not an API, so slower. No intraday.
from yahoofinancials import YahooFinancials
```

```python
ticker = 'MSFT' #create an object and perform functions
yahoo_financials = YahooFinancials(ticker)
data4 = yahoo_financials.get_historical_price_data("2018-04-24", "2020-04-24", "daily")
print(type(data4)) #data is a dictionary with JSON structure (nested dictionary)
data4["MSFT"]["prices"]
```

```python
all_tickers = ["AAPL","MSFT","CSCO","AMZN","INTC"]

# extracting stock data (historical close price) for the stocks identified
close_prices = pd.DataFrame() #create empty df
end_date = (dt.date.today()).strftime('%Y-%m-%d') #convert to string format time
beg_date = (dt.date.today()-dt.timedelta(1825)).strftime('%Y-%m-%d') #5 years of data
for ticker in all_tickers:
    yahoo_financials = YahooFinancials(ticker)
    json_obj = yahoo_financials.get_historical_price_data(beg_date,end_date,"daily")
    ohlv = json_obj[ticker]['prices'] #this is a list of dictionaries
    temp = pd.DataFrame(ohlv)[["formatted_date","adjclose"]] #converts to a df, taking just the formatted date and adj closing price
    temp.set_index("formatted_date",inplace=True) #makes formatted date as the index
    temp.dropna(inplace=True) #removes rows with missing values
    close_prices[ticker] = temp["adjclose"] #appends to empty df
close_prices.head(10)
```

```python
# extracting stock data (ohlcv) for the stocks identified
ohlv_dict = {}
end_date = (dt.date.today()).strftime('%Y-%m-%d')
beg_date = (dt.date.today()-dt.timedelta(1825)).strftime('%Y-%m-%d')
for ticker in all_tickers:
    yahoo_financials = YahooFinancials(ticker)
    json_obj = yahoo_financials.get_historical_price_data(beg_date,end_date,"daily")
    ohlv = json_obj[ticker]['prices']
    temp = pd.DataFrame(ohlv)[["formatted_date","adjclose","open","low","high","volume"]]
    temp.set_index("formatted_date",inplace=True)
    temp.dropna(inplace=True)
    ohlv_dict[ticker] = temp
```
